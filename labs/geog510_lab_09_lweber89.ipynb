## Ex. 3: Creating Marker Clusters

1. Create a marker cluster visualization from a GeoJSON file of building centroids:
   - URL: https://github.com/opengeos/datasets/releases/download/places/wa_building_centroids.geojson
   - Hint: Read the GeoJSON file using GeoPandas and add "latitude" and "longitude" columns to the GeoDataFrame.
2. Create circle markers for each building centroid using the `Map.add_circle_markers_from_xy()` method with the following styling:
   - Radius: 5
   - Outline color: "red"
   - Fill color: "yellow"
   - Fill opacity: 0.8

```python
import geopandas as gpd

url = "https://github.com/opengeos/datasets/releases/download/places/wa_building_centroids.geojson"

gdf = gpd.read_file(url)


#Add longitude and latitude to gdf

gdf['longitude'] = gdf.geometry.x
gdf['latitude'] = gdf.geometry.y


```

```python

m=leafmap.Map()

#Add satellite basemap (no labels)
m.add_basemap("SATELLITE")

#Add marker clusters based on modified gdf

m.add_marker_cluster(gdf, x="longitude", y="latitude", layer_name="Buildings")

m.zoom_to_gdf(gdf)

#Display map
m
```

```python
m = leafmap.Map(basemap="Esri.WorldImagery")
m.add_circle_markers_from_xy(gdf,
                              x="longitude",
                              y="latitude",
                              radius=5,
                              color="red",
                              fill_color="yellow",
                              opacity=.8
)

m.zoom_to_gdf(gdf)
m
```

## Ex. 4: Visualizing Vector Data

1. Visualize the building polygons GeoJSON file and style it with:

   - Outline color: "red"
   - No fill color
   - URL: https://github.com/opengeos/datasets/releases/download/places/wa_overture_buildings.geojson

2. Visualize the road polylines GeoJSON file and style it with:

   - Line color: "red"
   - Line width: 2
   - URL: https://github.com/opengeos/datasets/releases/download/places/las_vegas_roads.geojson

3. Create a choropleth map of county areas in the US:
   - URL: https://github.com/opengeos/datasets/releases/download/us/us_counties.geojson
   - Column: `CENSUSAREA`

```python
m = leafmap.Map(basemap="Esri.WorldImagery")

url = "https://github.com/opengeos/datasets/releases/download/places/wa_overture_buildings.geojson"

style = {
    "stroke": True,
    "color": "#ff0000",
    "weight": 2,
    "opacity": 1,
    "fill": True,
    "fillColor": "#ffffff",
    "fillOpacity": 0,
}

hover_style = {"stroke": True, "color": "#0000ff"}

m.add_geojson(url, style=style, hover_style=hover_style, zoom_to_layer=True)

m
```

```python
m = leafmap.Map(basemap="Esri.WorldImagery")

url = "https://github.com/opengeos/datasets/releases/download/places/las_vegas_roads.geojson"

style = {
    "stroke": True,
    "color": "#ff0000",
    "weight": 2,
    "opacity": 1,
}

hover_style = {"stroke": True, "color": "#00ff00"}

m.add_geojson(url, style=style, hover_style=hover_style, zoom_to_layer=True)

m
```

```python
m = leafmap.Map(center=[40, -100], zoom=4)

url = "https://github.com/opengeos/datasets/releases/download/us/us_counties.geojson"

m.add_data(
    url, column="CENSUSAREA", scheme="Quantiles", cmap="Blues", legend_title="Census Area"
)

m
```

## Ex. 5: Creating a Split Map

Create a split map to compare imagery of Libya before and after the 2023 flood event:

Pre-event imagery: https://github.com/opengeos/datasets/releases/download/raster/Libya-2023-07-01.tif

Post-event imagery: https://github.com/opengeos/datasets/releases/download/raster/Libya-2023-09-13.tif

```python
m = leafmap.Map()

#Assign pre- and post- event COG to variables

url_pre = "https://github.com/opengeos/datasets/releases/download/raster/Libya-2023-07-01.tif"
url_post = "https://github.com/opengeos/datasets/releases/download/raster/Libya-2023-09-13.tif"


#Populate left layer, right layer and label as appropriate

m.split_map(
    left_layer=url_pre, right_layer=url_post, left_label="Pre-Flood (07/01/2023)", right_label="Post-Flood (09/13/2023)"
)
m
```
