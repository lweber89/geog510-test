## Exercise 12: Extracting and Formatting Coordinates

- Given a string with the format `"latitude, longitude"` (e.g., `"40.7128N, 74.0060W"`), extract the numeric values of latitude and longitude.
- Convert these values to floats and remove the directional indicators (`N`, `S`, `E`, `W`).
- Format the coordinates into a `POINT` WKT string (e.g., `"POINT(-74.0060 40.7128)"`).

```python
#Initialize variable

geocoord_string = "40.7128N, 74.0060W"

#Initial unpack based on seperator
parts = geocoord_string.split(",")

#Strip unpacked components
lat_str = parts[0].strip()
lon_str = parts[1].strip()

print("String Unpacked")
print (lat_str)
print (lon_str)

#Remove N/W and convert to floats

print("\nStrip and Convert to floats")
lat_flt = float(lat_str[:-1]) 
lon_flt = float(lon_str[:-1]) 

print(lat_flt)
print(lon_flt)

#Create WKT representation

wkt_pt = f"POINT({lon_flt} {lat_flt})"
print("\n")
print(f"WKT Point:  {wkt_pt}")
```

## Exercise 14: String Normalization and Cleaning

## Exercise 15: Parsing and Extracting Address Information

- Given a string in the format `"Street, City, Country"` (e.g., `"123 Main St, Springfield, USA"`), write a function that parses the string into a dictionary with keys `street`, `city`, and `country`.
- The function should return a dictionary like `{"street": "123 Main St", "city": "Springfield", "country": "USA"}`.

```python
#Define function that takes in a string address and returns a dictionary

def address_dictionary_builder (address):

#Initialize empty dictionary

    address_dictionary = {
        "street": "",
        "city": "",
        "country": "",
        }

    #Split address on ","
    address_parts = address.split(",")

    #Assign address components to dictionary entries

    address_dictionary["street"] = address_parts[0].strip()
    address_dictionary["city"] = address_parts[1].strip()
    address_dictionary["country"] = address_parts[2].strip()
    
    #return completed dictionary
    
    return address_dictionary
```

```python
#Example usage

address = "123 Main St, Springfield, USA"
d = address_dictionary_builder(address)
print(d)
```

```python
#Create list of coordinates

coordinates = [
    (34.0522, -118.2437),  # Los Angeles
    (-33.8688, 151.2093),  # Sydney
    (40.7128, -74.0060),   # New York
    (-23.5505, -46.6333),  # São Paulo
    (0.0000, 32.4467),     # Entebbe (Equator)
    (51.5074, -0.1278),    # London
    (-34.6037, -58.3816),  # Buenos Aires
    (35.6895, 139.6917),   # Tokyo
    (-1.2921, 36.8219),    # Nairobi
    (48.8566, 2.3522),     # Paris
    (38.6270, -90.1994)    # St. Louis
]

#Loop through coordinates list and check for Northern or Southern Hemisphere - output result

for lat, lon in coordinates:
    if lat > 0:
        print(f"The coordinate {lat:.4f}, {lon:.4f} is in the Northern Hemisphere.")
    elif lat < 0:
        print(f"The coordinate {lat:.4f}, {lon:.4f} is in the Southern Hemisphere.")
    else:
        print(f"The coordinate {lat:.4f}, {lon:.4f} is on the equator!")
```

## Exercise 17: While Loops for Iterative Processing

- Create a list of coordinates (latitude, longitude).
- Write a `while` loop that continues to print each coordinate until it encounters a coordinate with a negative latitude.
- Stop the loop once this condition is met.

```python
#Create list of coordinates
coordinates = [
    (45.5231, -122.6765), # Portland
    (52.3676, 4.9041),   # Amsterdam
    (39.9042, 116.4074),  # Beijing
    (38.6270, -90.1994),  # St. Louis
    (-1.2921, 36.8219),   # Nairobi
    (40.7128, -74.0060),  # New York
    (-33.8688, 151.2093)  # Sydney
]

#Contruct while loop to iterate through coordinates and output until negative latitude is encountered

counter = 0

while counter < len(coordinates):
    lat, lon = coordinates[counter] #extract lat/lon from coord tuple

    if lat >= 0:
        print(f"{lat:.4f}, {lon:.4f}")
    else:
        break

    counter += 1
```

## Exercise 20: Generating and Analyzing Random Coordinates

- Write a program that generates random coordinates (latitude between [-90, 90] degrees and longitude between [-180, 180] degrees).
- Use a `while` loop to keep generating coordinates until a pair with both latitude and longitude greater than 50 is generated.
- Print each generated coordinate and the final coordinate that meets the condition.

```python
import random

while True:
    random_lat = random.uniform(-90, 90)  #create a random latitude
    random_lon = random.uniform(-180, 180) #create a random longitude

    #check if lat and lon are > 50, print output and break once both are > 50

    if random_lat <= 50 and random_lon <= 50:
        print(f"Coord: {random_lat:.4f}, {random_lon:.4f}")

    if random_lat > 50 and random_lon > 50:
        print(f"Final Coord: {random_lat:.4f}, {random_lon:.4f}")
        break
```
